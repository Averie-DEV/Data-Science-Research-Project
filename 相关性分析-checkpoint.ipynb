```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
billboard = pd.read_csv('step1 数据清洗\\billboard.csv')
billboard.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
taeyeon = pd.read_csv('step1 数据清洗\\taeyeon.csv')
taeyeon.head()
```

```python
import pandas as pd
from scipy import stats

# 假设billboard和taeyeon是已经加载的DataFrame
# 检查特征 'feature1' 是否在两个歌单中有显著差异
features = ['energy', 'key_', 'loudness','speechiness',
            'acousticness','instrumentalness','liveness',
            'valence', 'tempo']

for feature in features:
    print('testing on: '+str(feature))
# 首先检验数据的正态性
    print("Normality Test (Shapiro-Wilk):")
    print("Billboard:", stats.shapiro(billboard[feature]))
    print("Taeyeon:", stats.shapiro(taeyeon[feature]))

# 检验方差齐性
    print("Variance Equality Test (Levene's Test):")
    print(stats.levene(billboard[feature], taeyeon[feature]))

# 进行t检验
    t_stat, p_val = stats.ttest_ind(billboard[feature], taeyeon[feature], equal_var=True)
    print("T-test result:", t_stat, "P-value:", p_val)

# 根据P-value解释结果
    if p_val < 0.05:
        print("两个歌单在" +str(feature)+"上存在显著差异。")
    else:
        print("两个歌单在" +str(feature)+"上没有显著差异。")
```
